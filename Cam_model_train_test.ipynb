```python
import os
import shutil

image_folder = 'datasets/val/images'  # 이미지가 있는 폴더 경로
label_folder = 'datasets/val/val_labels'  # 레이블 파일이 있는 폴더 경로

# 이미지 폴더 내의 파일 목록 가져오기
image_files = [f for f in os.listdir(image_folder) if f.endswith('.jpg')]

for image_file in image_files:
    image_path = os.path.join(image_folder, image_file)
    label_file = image_file.replace('.jpg', '.txt')  # 이미지 파일 이름과 동일한 레이블 파일 이름 생성
    label_path = os.path.join(label_folder, label_file)

    # 이미지 파일과 레이블 파일을 함께 위치시키기
    if os.path.exists(label_path):
        new_label_path = os.path.join(image_folder, label_file)
        shutil.move(label_path, new_label_path)
        print(f'Moved {label_file} to {image_folder}')
    else:
        print(f'Label file {label_file} not found')

print('Process completed.')
```

```python
import os
import shutil

# Define the paths
data_dir = 'datasets'
train_dir = os.path.join(data_dir, 'train')
val_dir = os.path.join(data_dir, 'val')

# Create train_labels and val_labels directories if they don't exist
train_labels_dir = os.path.join(train_dir, 'train_labels')
val_labels_dir = os.path.join(val_dir, 'val_labels')
os.makedirs(train_labels_dir, exist_ok=True)
os.makedirs(val_labels_dir, exist_ok=True)

# Move label files to appropriate directories
for root, _, files in os.walk(train_dir):
    for file in files:
        if file.endswith('.txt'):
            label_file = os.path.join(root, file)
            dest_dir = train_labels_dir if 'train' in root else val_labels_dir
            shutil.move(label_file, os.path.join(dest_dir, file))

# Generate labels.cache
for root, _, files in os.walk(train_labels_dir):
    with open(os.path.join(root, 'labels.cache'), 'w') as f:
        for file in files:
            f.write(file.replace('.txt', '') + '\n')

# Print success message
print("Labels and labels.cache generated successfully.")
```
